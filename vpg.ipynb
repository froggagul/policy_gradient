## Vanilla Policy Gradient

```python
%load_ext autoreload
%autoreload 2

import gym
from torch.optim import Adam

from src.policy import MLPPolicy, VanillaPolicyGradient
from src.render import IpynbRenderer, AbstractRenderer

env = gym.make("CartPole-v1", render_mode="rgb_array")
renderer = IpynbRenderer()
obs_dim = env.observation_space.shape[0]
n_acts = env.action_space.n
model = MLPPolicy(
    [
        obs_dim,
        16,
        n_acts,
    ]
)

model.configure_optimizer(Adam, 0.01)

vpg = VanillaPolicyGradient(env, model, renderer)
vpg.train(epochs=100, batch_size=5000)
```

```python
vpg.test(1)
```
